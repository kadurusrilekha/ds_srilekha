```python
# ===========================================
# DATA SCIENCE ASSIGNMENT – WEB3 TRADING TEAM
# Candidate: your_name_here
# ===========================================

# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Step 2: Load datasets
trader_data = pd.read_csv('historical_data.csv')
sentiment_data = pd.read_csv('fear_greed_index[2].csv')

# Step 3: Preview data
print(trader_data.head())
print(sentiment_data.head())

# Step 4: Data Cleaning
trader_data.dropna(inplace=True)
sentiment_data.dropna(inplace=True)

# Convert timestamps to datetime and rename column
trader_data['Timestamp IST'] = pd.to_datetime(trader_data['Timestamp IST'], errors='coerce', dayfirst=True)
trader_data.rename(columns={'Timestamp IST': 'time'}, inplace=True)
sentiment_data['date'] = pd.to_datetime(sentiment_data['date'])

# Create 'date' column in trader_data
trader_data['date'] = pd.to_datetime(trader_data['time'].dt.date)

# Debug prints to verify merge keys
print(trader_data.columns)
print(sentiment_data.columns)
print(trader_data['date'].head())
print(sentiment_data['date'].head())
print('Trader duplicates:', trader_data['date'].duplicated().any())
print('Sentiment duplicates:', sentiment_data['date'].duplicated().any())

# Sort before merging to avoid subtle errors
trader_data = trader_data.sort_values('date')
sentiment_data = sentiment_data.sort_values('date')

# Step 5: Merge datasets by 'date'
merged = pd.merge(trader_data, sentiment_data, on='date', how='inner')

# Step 6: Exploratory Data Analysis
plt.figure(figsize=(10, 5))
sns.countplot(data=sentiment_data, x='classification')
plt.title('Market Sentiment Counts (Fear vs Greed)')
os.makedirs('/content/outputs', exist_ok=True)
plt.savefig('/content/outputs/sentiment_overview.png')
plt.close()

# Step 7: Correlation Analysis
metrics = merged[['Closed PnL', 'value']]
corr = metrics.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.savefig('/content/outputs/correlation_matrix.png')
plt.close()

# Step 8: Insights
summary = merged.groupby('classification')['Closed PnL'].mean()
print("Average PnL by Sentiment:")
print(summary)

# Step 9: Export processed data
os.makedirs('/content/csv_files', exist_ok=True)
merged.to_csv('/content/csv_files/merged_data.csv', index=False)
```
